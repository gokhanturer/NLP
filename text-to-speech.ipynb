```python
from gtts import gTTS
import os
```

```python
#burada tr yerine hangi dili yazarsak text olarak verilen girdiyi o aksanla seslendiriyor

mytext="""
Yapay zeka sistemlerinin iniş ve çıkış dönemleri olduğunu artık hepimiz biliyoruz. 
Teknoloji geliştikçe yapay zeka alanına yapılan yatırım farklı evrelere dönüşerek bir anda karşımıza çıkıyor. 
Artık hayatımızın her alanında irili ufaklı yapay zeka sistemlerinin çıkmasını bu evrelere borçluyuz.
"""
output=gTTS(text=mytext,lang='tr',slow=False)
output.save('output_tr.mp3')
```

```python
from IPython.display import Audio
audio=Audio(filename='output_tr.mp3')
audio
```
