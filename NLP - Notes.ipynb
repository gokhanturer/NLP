# Basic Conceptes

- **Corpora:** Birden fazla datasetin bulunduğu kütüphane
- **Corpus:** Elimizdeki dataset bizim corpusumuz
- **Document:** Document corpusdaki herbir satır. Bir kelimeden de oluşabilir, bir kitapdan da oluşabilir
- **Token:** Documentlerimizden anlamlı veriler çıkarabilmek için küçük paraçlara bölünmüş hali.  Sentence ya da word token olabilir.

# NLP Preprocess

## 1 - Lowercase sentence

- Bütün cümleler küçük harfe çevrilir
- Eğer aynı kelime hem küçük hem büyük harfle başlıyorsa CountVectorizer ve TF-IDF bu kelimeleri farklı iki kelime alacaktır

## 2 - Tokenization

- Bir cümleyi, paragrafı, söz öbeğini veya tüm metni küçük parçalara bölme işlemidir.
- Her bir kelime veya cümle token kabul edilebilir. 
- **Word token daha yagın kullanılmaktadır**



## 3-Removing Punctuation and Numbers

- Metinden string dşındaki karakterlerin ve numerik değerlerin çıkarılmasıdır.
- ML modeller için uygulanır, DL modeller için uygulamaya gerek yoktur.
- Count vectrizer veya TF-IDF için gereklidir

## 4 - Removing Stopwords

- Bir dilde çoklukla kullanılan ve eğitime çok katkısı olmayan kelimelerdir
- Model kurulmadan önce bu kelimeler metinden çıkarılır
- Bu kelimeler metindeki diğer kelimelerden çok kullanıldığı için gürültü oluşturur 
- **Ancak dugu analizi(sentiment) yapılan işlemlerde olumsuz anlam katan kelimeler(yardımcı fiiller) atılmaz**



**Some Turkish language stopwords**



## 5 - Stemming - Lemmatization

- **Stemming :** Bir kelimenin en yalın halini(kökünü) bulmaktır. 
- Bazen anlamsız kelimeler üretir, çok kullanışlı değildir
- **Lemmatization :** Bir kelimenin anlamlı kökünü bulmaktır.
- Bulunan kelimenin sözlükte bir karşılığı vardır
- Eğer kelime bulunduğu halde anlamlı bir kelime ise değiştirmez.
- Çekim eklerini atar yapım ekleri kalır(Türkçe karşılığı)

## 6- Vectorization

### a - Count Vectors(CountVectorizer)



- Count Vectorizer D1 ve D2'yi birleştirip bütün uniqe kelimeleri tespit ediyor.
- Bu kelimeleri alfabetik sıraya dizerek feature olarak alıyor
- Daha sonra herbir unique kelimenin sayısını buluyor






- **Burda bir handikap var !!!** <br>
<br>
- CountVectorizer aynı sayıda olan kelimelere eşit muamelede bulunuyor
- Hangi kelime daha önemli bunu bilemiyoruz. Sayısı fazla olana daha fazla ağırlık veriyor
- CountVectorizer bir tokenin document içerisindeki önemini veriyor ancak o tokenin corpus içerisindeki önemine ait bize bir insight sağlamıyor

### b- TF-IDF Vectors

- CountVectorizer'daki handikaptan dolayı TF-IDF'i kullanıyoruz



- TF-IDF Vectorization bir tokenin hem document içerisindeki hemde corpus içersindeki önemini veriyor
- **Term Frequency(TF) :** Bir kelimenin documentdeki sayısı / documentdeki tüm kelime sayısı 
- **Inverse Document Frequency :** Bir keliemenin geçtiği document sayısı / corpusdaki tüm döküman sayısı.
- Inverse ile bu işlemin tersini alıyoruz
- Çünkü 0-1 arasındaki sayıların logaritması negatiftir. 
- Örn log(1/5) negatiftir. Ancak log(5/1)'i hesaplarsak bu aynı değerin pozitif sonucudur<br>
<br>



### c - Word Embedding

- CountVectorizer ve TF-IDF de vektörlerimizin boyutu çok büyük(unique tokenlerin sayısı kadar) ve çoğu değerler 0 lardan oluşuyor
- CountVectorizer ve TF-IDF kelimenlerin anlamsal bütünlüğünü tespit edemediği için embedding modeller kullanılır
- Word Embedding tokenlar arası anlam bütünlüğünün ne kadar olduğunu cosine similarity ile ölçer
- Cosine similarity iki vektör arsındaki açının cosinüs değeridir. Cos(0) = 1
- Cosine similarity değeri 1 yaklaştıkça anlam ilişkisi o kadar artar.



 - **Embedding Layer :** DL modellerinde manuel oluşturulan embedding modelidir
 - **Word2Vec :** Google tarafından geliştirilen bir modeldir. Verilen metindeki kelimelerin tamamını tarayarak herbir kelimenin  öncesi ve sonrasındaki kelimelere bakarak anlamsal ilişki bulur
     - **Skip-Gram Model :** Skip-gramda girdi hedef kelime iken çıktılar ise hedef olarak verilen kelimenin etrafındaki kelimelerdir. Girdiler ile çıktıları olasılıksal olarak birbirine benzeterek anlamsal olarak en uygun şekilde temsil etmek hedeflenmektedir.Küçük corpuslarda iyi sonuçlar verir.
     
     - **Continuous Bag Of Words(CBOW MOdel):** Skip-gram’a çok benzer bir yaklaşımdır. Aralarındaki tek fark girişlerle çıkışların yer değişmiş olmasıdır. Buradaki fikir bir kelimenin etrafındaki kelimeler verildiğinde hangi kelimenin bu kelimeler içinde görülme olasılığının en yüksek olduğunu bilmek istemesidir. Büyük corpuslarda daha iyi sonuçlar verir.
     



     
 - **Global Vectors(Glove) :** Word2Vec'den farklı olarak birlikte kullanılan kelimeler corpusda ne kadar kullanılmış istatiksel hesap yapıyor ve vektör değerlerini belirliyor. Word2Vec bütün kelimeleri sayıyor birlikye kullanılan kelimeleri buldukça vektör değeri değişiyor
 - **FastText :** 2016 yılında Facebook tarafından geliştirilmiş Word2Vec’in bir uzantısıdır. Tek tek kelimeleri yapay sinir ağına girdi olarak vermek yerine kelimeleri birkaç harf bazlı “n-gram” halinde parçalar.
 - **Embeddings from Language Models(ELMo) :** Sesteş kelimelerin anlam farklılıklarını yakalayabiliyor. Word2Vec ve Glove tek yönlü modeller olduğu için sesteş kelimelere aynı vektör değerini veriyor. ELMo birectional modeldir
 - **Bidirectional Encoder Representations from Transformers(BERT):** Diğer modellerden farklı olarak cümleyi hem soldan sağa hem de sağdan sola olarak değerlendiriyor. Bu sayede anlamı ve kelimelerin birbiriyle olan ilişkileri daha iyi çıkarmayı planlıyor ve sonuçlarda da bunun karşılığını alıyor
 
 


# NLP with Deep Learning

## Recurrent Neural Network (RNN)

- RNN'ler, genellikle bir sonraki adımı tahmin etmek için kullanılan bir tür Derin Öğrenme yapısıdır. Onlarla diğer derin öğrenme yapıları arasındaki en büyük fark, hatırlamalarıdır. Diğer bir fark ise, diğer sinir ağlarında her giriş diğerinden bağımsız iken, RNN'lerde girişler birbiriyle ilişkilidir. RNN'ler, bir sonraki adımı takip etmek için girdiler arasında ilişkilendirmeler yapar ve eğitilirken tüm çağrışımlarını hatırlar.

- Ancak, vanishing ve exploding gradyanlar, RNN geri yayılımı sırasında ortaya çıkan iki sık sorundur. Vanishing gradyan probleminde gradyan değerleri hızla düşer ve eğitim durur. Model, uzun cümlelerin başı ve sonu arasındaki ilişkileri yakalayamaz. Ancak exploding gradyanlarda gradyan değerleri katlanarak sonsuza kadar büyür ve kararsız süreç nedeniyle model ağırlık değerleri NaN olur. Model, eğitim verilerinden hiçbir şey öğrenemez.

 - RNN sıralı verilerin veya time series verilerinde başarılı sonuçlar verir
 - RNN de bir nöronun çıktısı yeni veri ile tekrar girdi olarak verilir
 - ANN veya CNN gibi değildir, hazfızası vardır.
 - NLP özelinde RNN'lere cümle tokenlarını cümledeki sırasına göre vermek gerekir
 
 

Vektörler bir sinir ağından geçtiğinde, çeşitli matematik işlemleri sonucunda bir takım değişikliklere uğrarlar. Bu işlemler sonucunda bazı değerlerin çok büyük, bazı değerlerin ise önemsiz hale gelmemesi için tanh aktivasyonu kullanılır. Böylece tüm vektör değerleri -1 ile +1 değerleri arasında kalır.

- RNN lerde aktivasyon fonksiyonu olarak tanh fonksiyonu kullanıyoruz
- Tanh fonksiyonu girdi olarak ne alursa alsın çıktı olarak -1 ile +1 rasında değerler üretir.
- Vektör değerlerimin ağırlıklarının bozulmaması için tanh fonksiyonu kullanıyoruz



- **RNN Sequential Memory'de hafıza problemi karşımıza çıkmaktadır**
- **Şekilden de anlaşılacağı üzere başlangıçtan son tokene doğru gidildikçe başlangıçtaki tokenlar unutulmaktadır**<br>
<br>




## Long Short-Term Memory (LSTM)

- RNN lerde cümle verildiğinde sona geldiğinde cümlenin başı unutuluyordu
- Bu problemi gidermek için LSTM ce GRU modelleri geliştirilmiştir<br>
<br>

- LSTM in RNN den farklı cell state olmasıdır



- **Forget Gate :** Bu kapı, bilgilerin saklanıp saklanmayacağını belirler. Bu işlem için sigmoid fonksiyonu kullanılır. Değer 0'a ne kadar yakınsa unutulma olasılığı o kadar yüksek ve 1'e ne kadar yakınsa akılda kalma olasılığı o kadar yüksek olacaktır. Cell State'in uzun süre hafızasından çıkarılması gereken cümle kalıplarını belirliyor
- **Input Gate :** İlk aşamada, gizli durumdaki mevcut giriş, sigmoid işlevine ve ardından tanh işlevine geçirilir. Sigmoid fonksiyonundan elde edilen sonuçlar 0 ile 1 arasında olacağından tanh fonksiyonundan gelen bilgilerin hücre durumuna eklenip eklenmeyeceğini belirler. Cell State'e ilave yapılıyor
- **Cell State :** Hücre durumu, dizi boyunca ilgili bilgileri taşır. Böylece önceki bilgiler unutulmamakta ve model daha doğru tahminler yapabilmektedir. Bu nedenle, bunu ağın hafızası olarak düşünebilirsiniz.

**Bunu bir chatbot örneği ile açıklayalım**
- Başlangıçta hidden state 0 lardan oluşan bir vektördür
- İlk token gelince(what) modelim öncesinde chatbot verisi ile eğitildiği için soru kalıplarının çok önemli olduğunu biliyor
- "what" tokenini hemen cell state'e gönderiyor. Forget Gate de bir işlem yapılmıyor
- Daha sonra "what" tokeni input gate geliyor. Önce sigmoid sonra da tanh fonksiyonundan geçiyor
- Sigmoidde çıkan sonuç 1 ise "what" tokeni cell state ekleniyor.
- Son olarak hidden state ve cell state çıktıları tekrar girdi olarak veriliyor
- Bu işlemler tüm tokenlar boyunca tekrarlanıyor
- En son hidden state ile cell state outpu kapısı ile birleştiriliyor

- Yeni bir soru kalıbı geldiğinde(How old are you) "what .. " cümlesi cell stateden çıkarılıyor

## Gated Recurrent Units (GRU)



- LSTM ile oldukça benzerdir. İki gate vardır; Update Gate, Reset Gate
- LSTM de işlemler daha uzun sürmektedir, GRU bu konuda daha hızlıdır
- GRU'da Hidden State ve Cell State tek bir katmanda birleştirilmiştir
- **Update Gate:** LSTM'deki forget ve input kapıları gibi hangi bilgilerin atılacağına ve hangi bilgilerin dahil edileceğine karar verir.

- **Reset gate:** Diğer adımlardan gelen bilgilerin ne kadarının unutulacağına karar veren kapıdır.

- GRU, LSTM'den daha hızlı olmasına rağmen, hem LSTM hem de GRU çok iyi sonuçlar veriyor. 
- İkisini de deneyip karar vermek size kalmış.
